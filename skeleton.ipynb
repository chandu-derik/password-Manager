```python
from abc import ABC, abstractmethod
class Password(ABC):
    
    @abstractmethod
    def SignUp(self):
        pass
    @abstractmethod
    def login(self):
        pass
    @abstractmethod
    def random_password(self):
        pass
    @abstractmethod
    def User_preferences(self):
        pass
    @abstractmethod
    def File_storage(self):
        pass
    @abstractmethod
    def Menu(self):
        pass
    @abstractmethod
    def User_preference_password(self):
        pass
    @abstractmethod
    def user_menu(self):
        pass
```

```python
import os
import random
import string
```

```python
def load_users(filename="Users_list.txt"):
    users={}
    if os.path.exists(filename):
        with open(filename,'r') as file:
            for line in file:
                line=line.strip()
                if line:
                    username,password=line.split(",",1)
                    users[username]=password
    return users
```

```python
def save_users(users,filename="Users_list.txt"):
    with open(filename,'w') as file:
        for username,password in users.items():
            file.write(f"{username},{password}\n")
```

```python
def clear_screen():
    os.system("cls" if os.name == "nt" else "clear")
```

```python
class passwordGenerate(Password):

    def __init__(self):
        self.current_user=None
        self.store={}
    
    def SignUp(self):
        users=load_users()
        username=input("Enter your name:").lower()

        if username in users:
            print("User already exists ! please Login")
            return False

        password=input("Generate your password:")
        users[username]=password
        save_users(users)
        print("Signup sucessfull ! you can now log-in.")
        return True 
        
    def login(self):
        users=load_users()
        username=input("Enter your name : ").lower()
        password=input("Enter your password: ")

        if username in users and users[username]==password:
            print("Login sucessfull !")
            self.current_user=username
            return True 
        else:
            print("Invalid password or username")
            return False
    def Menu(self):
        self.main_menu()

    def main_menu(self):
        while True:
            print("1)signup")
            print("2)Login")
            print("3)Exit")
            
            choice=input("choose the number from the above:")
            if choice=="1":
                self.SignUp()
            elif choice=="2":
                if self.login():
                    self.user_menu()
                
            elif choice=="3":
                print("You are out...BYE!!!!!")
                break
            else:
                print("Please enter a valid number")
    def user_menu(self):
        while True:
            print("1)Generate Random Password")
            print("2)Generate special password")
            print("3)view saved Passwords")
            print("4)Delete the passwords")
            print("5)Exit")
            option=input("Choose the options :")

            if option=="1":
                clear_screen()
                password=self.random_password()
                if password:
                    print("Here is your random password",password)
                    if password:
                        website=input("Enter the website name :").strip().lower()
                        self.File_storage(website, password)
                        print("Password saved successfully!")
                input("\nPress Enter to continue...")
            elif option=="2":
                clear_screen()
                values=self.User_preferences()
                if values:
                    password=self.User_preference_password(*values)
                    print("Here is your special password :",password)
                    if password:
                        website=input("Enter the website name :").strip().lower()
                        self.File_storage(website, password)
                        print("Password saved successfully!")
                input("\nPress Enter to continue...")
            elif option=="3":
                clear_screen()
                self.view_passwords()
                input("\nPress Enter to continue...")
            elif option=="4":
                clear_screen()
                self.delete_password_by_number()
                input("\nPress Enter to continue...")
            elif option=="5":
                clear_screen()
                print("Logging Out.....")
                return
            else:
                print("choose the valid option")
                break
            
    def random_password(self):
        length=int(input("Enter the password length : "))
        
        if length < 4:
            print("length must be greater than 3 to include all the characters")
            return
        while True: 
            upper=random.choice(string.ascii_uppercase)
            lower=random.choice(string.ascii_lowercase)
            digit=random.choice(string.digits)
            special=random.choice(string.punctuation)

            password_length = length-4
            all_characters=string.ascii_lowercase+string.digits+string.ascii_uppercase+ "@"
            remaining_password=[random.choice(all_characters) for i in range(password_length)]

            password_list=[upper,lower,digit,special] + remaining_password
            random.shuffle(password_list)

            password="".join(password_list)
            print(f"Your Random password : {password}")
            choice = input("\n1) Keep this password\n2) Generate another\nChoose: ").strip()
            if choice=="1":
                return password
                break
            elif choice=="2":
                continue
            else:
                print("Invalid choice")

            

    def view_passwords(self):
        if not os.path.exists("passwords.txt"):
            print("No saved passwords")
            return []

        user_passwords = []

        with open("passwords.txt", "r") as file:
            for line in file:
                username, website, password = line.strip().split(",",2)
                if username == self.current_user:
                    user_passwords.append((website, password))

        if not user_passwords:
            print("No saved passwords found")
            return []

        print("\nSaved Passwords:")
        for idx, (website, _) in enumerate(user_passwords, start=1):
            print(f"{idx}. {website}")

        return user_passwords

    def delete_password_by_number(self):
        print("\n--- DELETE PASSWORD MODE ---")

        user_passwords = self.view_passwords()

        print("DEBUG: passwords found =", len(user_passwords))

        if len(user_passwords) == 0:
            print("Nothing to delete")
            return

        print("\nEnter the number of the password you want to delete.")
        choice = input("Number: ")

        if not choice.isdigit():
             print("Invalid input")
             return

        choice = int(choice)

        if choice < 1 or choice > len(user_passwords):
             print("Number out of range")
             return

        website_to_delete = user_passwords[choice - 1][0]

        confirm = input(f"Are you sure you want to delete the password for '{website_to_delete}'? (yes/no): ").strip().lower()

        if confirm != "yes":
            print("Deletion cancelled")
            return

        new_lines = []
        with open("passwords.txt", "r") as file:
            for line in file:
                username, website, password = line.strip().split(",", 2)
                if not (username == self.current_user and website == website_to_delete):
                    new_lines.append(line)

        with open("passwords.txt", "w") as file:
            file.writelines(new_lines)

        print(f"\n Password has been Deleted'{website_to_delete}'")

    def User_preferences(self):
        try:
            length = int(input("Enter the password length : "))
        except ValueError:
            print("Please enter a valid number")
            return None
        Upper=input("Include Uppercase [yes/no] ? :").strip().lower()=="yes"
        Lower=input("Include Lowercase [yes/no] ? :").lower().strip()=="yes"
        Digit=input("Include Digit [yes/no] ? :").lower().strip()=="yes"
        special=input("Include Specail Characters [yes/no] ? :").lower().strip()=="yes"

        word=input("Enter a word to include (optional) :")
        
        if not(Upper or Lower or Digit or special):
            Upper=Lower=Digit=special =True

        return length, Upper, Lower, Digit, special, word
        
    def User_preference_password(self, length, Upper, Lower, Digit, special, word):
        if word and length < len(word):
            print("Required password length is less than the word")
            return None
            
        remaining_length = length - len(word)
        password_freq = ""
        required_password = []
        
        if Upper:
            required_password.append(random.choice(string.ascii_uppercase))
            password_freq += string.ascii_uppercase
        if Lower:
            required_password.append(random.choice(string.ascii_lowercase))
            password_freq += string.ascii_lowercase
        if Digit:
            required_password.append(random.choice(string.digits))
            password_freq += string.digits
        if special:
            required_password.append(random.choice(string.punctuation))
            password_freq += string.punctuation

        if not password_freq:
            password_freq = string.ascii_letters + string.digits + string.punctuation

        if remaining_length < len(required_password):
            print("The password cannot be created because of length mismatch!")
            return None

    
        while True:
           
           remaining_chars = required_password[:]

           for _ in range(remaining_length - len(required_password)):
               remaining_chars.append(random.choice(password_freq))

           random.shuffle(remaining_chars)

           final_password = word + "".join(remaining_chars)
           print(f"\nHere is your special password : {final_password}")
            
           again = input("Do you want another special password [yes/no] ? : ").strip().lower()
            
           if again != "yes":
              break
        return final_password
    
    def File_storage(self, website, password):
        with open("passwords.txt", "a") as file:
            file.write(f"{self.current_user},{website},{password}\n")
```
